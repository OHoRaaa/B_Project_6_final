# 크롤링

```python
title = []
brand = []
pnum = []
season = []
gender = []
view_num = []
sale_num = []
good = []
star = []
review_num = []
tag = []
price = []

cnt = 29317

for page in range(330, 350):
    
    url = f'https://www.musinsa.com/category/003?d_cat_cd=003&brand=&rate=&page_kind=search&list_kind=small&sort=pop&sub_sort=&page={page}&display_cnt=90&sale_goods=&group_sale=&kids=N&ex_soldout=&color=&price1=&price2=&exclusive_yn=&shoeSizeOption=&tags=&campaign_id=&timesale_yn=&q=&includeKeywords=&measure='
    
    option = webdriver.ChromeOptions()
    option.add_argument('--no-sandbox')
    option.add_argument("disable-gpu")
    option.add_argument("--lang=ko_KR")
    option.add_argument('user-agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36')
    option.add_argument('--disable-blink-features=AutomationControlled')
    option.add_argument("--disable-extensions")
    option.add_experimental_option('useAutomationExtension', False)
    option.add_experimental_option("excludeSwitches", ["enable-automation"])
    driver = webdriver.Chrome('D:\B_Project6_Final\chromedriver.exe', options=option)
    driver.get(url)


    driver.maximize_window()
    
    element1 = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "searchList")))
    
    product_list = driver.find_elements_by_css_selector('#searchList > li')
    
#     pro_title = soup_t.select('span.product_title > em')
#     print(pro_title)
    
    for num in range(1, len(product_list) + 1):
        
        try:

            element2 = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "searchList")))

            driver.find_element_by_css_selector(f'#searchList > li:nth-child({num}) > div.li_inner > div.article_info > p.list_info > a').click()

            time.sleep(1.5)

            image = driver.find_elements_by_css_selector("div.product-img > img")[0]
            image = image.get_attribute('src')

            html = driver.page_source
            soup_t = BeautifulSoup(html, 'html.parser')

            pro_title = soup_t.select('span.product_title > em')
            print(pro_title)
            pro_brand = soup_t.select('ul.product_article > li > p > strong')
            pro_pnum = soup_t.select('ul.product_article > li > p > strong')
            pro_season = soup_t.select('ul.product_article > li > p > strong')
            pro_gender = soup_t.select('span.txt_gender')
            pro_view_num = soup_t.select('#pageview_1m')
            pro_sale_num = soup_t.select('#sales_1y_qty')
            pro_good = soup_t.select('span.prd_like_cnt')
            pro_star = soup_t.select('span.prd-score__rating')
            pro_review_num = soup_t.select('span.prd-score__review-count')
            pro_tag = soup_t.select('li.article-tag-list.list')
            pro_price = soup_t.select('#list_price')

            pro_li = [pro_title, pro_brand, pro_pnum, pro_season, pro_gender, pro_view_num, pro_sale_num, pro_good, pro_star,\
                      pro_review_num, pro_tag, pro_price]

            if len(pro_li[0]) >= 1:        
                title.append(pro_title[0].text)
                print(pro_title[0].text)
            else:
                title.append('')

            if len(pro_li[1]) >= 1:
                brand.append(pro_brand[0].text.split('/')[0].strip(' '))
                print(pro_brand[0].text.split('/')[0].strip(' '))
            else:
                brand.append('')

            if len(pro_li[2]) >= 1:
                pnum.append(pro_pnum[0].text.split('/')[1].strip(' '))
                print(pro_pnum[0].text.split('/')[1].strip(' '))
            else:
                pnum.append('')

            if len(pro_li[3]) >= 1:
                season.append(pro_season[1].text.replace(' ', '').replace('\n', ''))
                print(pro_season[1].text.replace(' ', '').replace('\n', ''))
            else:
                season.append('')

            if len(pro_li[4]) >= 1:
                gender.append(pro_gender[0].text.replace('\n', ' '))
                print(pro_gender[0].text.replace('\n', ' '))
            else:
                gender.append('')

            if len(pro_li[5]) >= 1:
                view_num.append(pro_view_num[0].text)
                print(pro_view_num[0].text)
            else:
                view_num.append('')

            if len(pro_li[6]) >= 1:
                sale_num.append(pro_sale_num[0].text)
                print(pro_sale_num[0].text)
            else:
                sale_num.append('')

            if len(pro_li[7]) >= 1:
                good.append(pro_good[0].text)
                print(pro_good[0].text)
            else:
                good.append('')

            if len(pro_li[8]) >= 1:
                star.append(pro_star[0].text)
                print(pro_star[0].text)
            else:
                star.append('')

            if len(pro_li[9]) >= 1:
                review_num.append(pro_review_num[0].text)
                print(pro_review_num[0].text)
            else:
                review_num.append('')

            if len(pro_li[10]) >= 1:
                tag.append(pro_tag[0].text.replace('\n', ' '))
                print(pro_tag[0].text.replace('\n', ' '))
            else:
                tag.append('')

            if len(pro_li[11]) >= 1:
                price.append(pro_price[0].text)
                print(pro_price[0].text)
            else:
                price.append('')


            df = pd.DataFrame({'제목' : title, '브랜드' : brand, '품번' : pnum, '시즌' : season, '성별' : gender,\
                            '조회수(1개월)' : view_num, '누적판매(1년)' : sale_num, '좋아요' : good,\
                          '별점' : star, '후기개수' : review_num, '태그' : tag, '가격' : price})
            
            if num == 90:
                df.to_csv(f'바지{page}.csv')

            pro_name = pro_title[0].text

            cnt += 1

            urllib.request.urlretrieve(image, "image_data/신발/" + re.sub(r'[^\w\s]', '', pro_name) + ' ' + str(cnt) + ".jpg")

            print('')
            print('개수: ', num, '개')
            print('=' * 50)

            driver.back()
            time.sleep(1.5)
        
        except:
            driver.back()
            time.sleep(1.5)


    print('')
    print('')
    print('='* 70, page, '=' * 70)

        
```
