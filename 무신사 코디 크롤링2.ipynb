#  코디 크롤링

```python
cnt = 0
img_cnt = 0

num_li = []
title_li = []
date_li = []
info_li = []


for page in range(1, 101):
    url = f'https://www.musinsa.com/app/styles/lists?use_yn_360=&style_type=&brand=&model=&tag_no=&max_rt=&min_rt=&display_cnt=60&list_kind=big&sort=date&page={page}'
    driver = webdriver.Chrome()
    driver.get(url)
    driver.maximize_window()
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    nums = len(soup.select('strong.style-list-information__title'))
    
    for num in range(1, nums + 1):
        pro_img = soup.select(f'body > div.wrap > div.right_area > form > div.right_contents.hover_box > div > ul > li:nth-child({num}) > div.style-list-item__thumbnail > a > div > img')[0]
        piu = "http:" + pro_img['src']
        cnt += 1
        urllib.request.urlretrieve(piu, f"D:\B_Project6_Final\image_data\코디2\\{cnt}.jpg")
        

        driver.find_element_by_css_selector(f'body > div.wrap > div.right_area > form > div.right_contents.hover_box > div > ul > li:nth-child({num}) > div.style-list-information > a > strong').click()
        try:
            time.sleep(0.5)
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')

            title_li.append(soup.select('#style_info > h2')[0].text)
            date_li.append(soup.select('p.styling_date')[0].text.split('|')[0])
            info_li.append(soup.select('p.styling_txt')[0].text)

            img_cnt += 1
            num_li.append(f'{img_cnt}.jpg')

            df = pd.DataFrame({'사진' : num_li, '룩' : info_li, '제목' : title_li, '날짜' : date_li})

            driver.back()
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            time.sleep(1.5)
        except:
            
            img_cnt += 1
            driver.back()
            time.sleep(1.5)
```
