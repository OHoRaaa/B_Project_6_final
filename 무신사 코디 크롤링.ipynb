# 코디 크롤링

```python
cnt = 0
img_cnt = 0

info_li = []
title_li = []
date_li = []
num_li = []


for page in range(1, 386):
    url = f'https://www.musinsa.com/app/styles/lists?use_yn_360=&style_type=&brand=&model=&tag_no=&max_rt=&min_rt=&display_cnt=60&list_kind=big&sort=date&page={page}'
    
    res = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
    
    soup = BeautifulSoup(res.text, 'html.parser')
    
    product_img_url = soup.select('div.style-list-thumbnail > img')
    
    info = soup.select('span.style-list-information__text')
    title = soup.select('strong.style-list-information__title')
    length = len(info)
    
    for img in product_img_url:                        # 이미지 저장
        piu = "http:" + img['src']
        cnt += 1
        urllib.request.urlretrieve(piu, f"D:\B_Project6_Final\image_data\코디\\{cnt}.jpg")
    
    for num in range(length):                         # 데이터 프레임
        info_li.append(info[num].text)
        title_li.append(title[num].text)

        if soup.select(f'body > div.wrap > div.right_area > form > div.right_contents.hover_box > div > ul > li:nth-child({num+1}) > div.post-information > span:nth-child(1)')[0].text == 'N':
            date_li.append(soup.select(f'body > div.wrap > div.right_area > form > div.right_contents.hover_box > div > ul > li:nth-child({num+1}) > div.post-information > span:nth-child(2)')[0].text)
        else:
            date_li.append(soup.select(f'body > div.wrap > div.right_area > form > div.right_contents.hover_box > div > ul > li:nth-child({num+1}) > div.post-information > span:nth-child(1)')[0].text)
    #     date_li.append(date[num + cnt].text)
        img_cnt += 1
        num_li.append(f'{img_cnt}.jpg')
        df = pd.DataFrame({'사진' : num_li, '룩' : info_li, '제목' : title_li, '날짜' : date_li})
        
    print(page, '페이지', ' 완료')
```
